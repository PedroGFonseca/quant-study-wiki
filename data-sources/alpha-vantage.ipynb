## Historical time series data - Intraday

```python
from alpha_vantage.timeseries import TimeSeries
```

```python
ts = TimeSeries(key=api_key, output_format='pandas')
data, meta_data = ts.get_intraday(symbol='MSFT',interval='1min', outputsize='full')
```

```python
data = data.reset_index()
data['date'] = pd.to_datetime(data.reset_index().date).values
data = data.set_index('date')
```

## Technical indicators

```python
from alpha_vantage.techindicators import TechIndicators
```

```python
ti = TechIndicators(key=api_key, output_format='pandas')
data, meta_data = ti.get_bbands(symbol='MSFT', interval='1min', time_period=60)
data.plot()
plt.title('BBbands indicator for  MSFT stock (1 min)')
plt.show()
```
